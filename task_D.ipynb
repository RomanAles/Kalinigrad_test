```python
import numpy as np
import pandas as pd
from scipy.optimize import minimize
```

```python
data = pd.read_csv('data.csv',header=None)
data
```

```python
x_data = data[0].values  # Первое значение в каждой строке
f_data = data[1].values  # Второе значение в каждой строке
```

```python
# Определение функции ошибки
def error_function(params):
    a, b, c = params
    # Предсказанное значение функции f(x)
    predicted_f = ((a * np.sin(x_data) + b * np.log(x_data)) ** 2 + c * x_data ** 2)
    # Сумма квадратов ошибок
    error = np.sum((predicted_f - f_data) ** 2)
    return error
```

```python

# Начальное приближение для параметров a, b, c
initial_guess = [1.0, 1.0, 1.0]

# Ограничения: a, b, c > 0
bounds = [(0, None), (0, None), (0, None)]

# Минимизация функции ошибки
result = minimize(error_function, initial_guess, bounds=bounds, method='L-BFGS-B')

# Получение оптимальных значений a, b, c
a_opt, b_opt, c_opt = result.x

# Вывод результатов с точностью до двух знаков после запятой
print(f"{a_opt:.2f} {b_opt:.2f} {c_opt:.2f}")
```
